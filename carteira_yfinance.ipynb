# Criando uma carteira

```python
carteira = ['PETR4','VALE3','ABEV3','WEGE3','EMBR3','RADL3','CVCB3','CIEL3','OIBR3','KNRI11']
def retorna_ativos(ativos,inicio,fim):
  df = pd.DataFrame()
  for i in ativos:
    ticker = str(i) + '.SA'
    df[i] = yf.download(ticker, start = inicio, end = fim)['Adj Close']

  return df
```

```python
precos = retorna_ativos(carteira,inicio,fim)
```

```python
precos.head()
```

### Quantos papeis

```python
primeira = precos.iloc[0]
primeira
```

```python
comprados = round(1000/primeira) # saber quantos papeis compramos com 10.000 reais
```

# Calcular patrimono liquido

```python
PL = precos * comprados
PL
```

```python
PL['Minha Carteira'] = PL.iloc[:].sum( axis = 1)
```

```python
PL.head()
```

```python
PL.tail()
```

# Comparando com IBOV

```python
ibov = yf.download('^BVSP', start = inicio, end = fim)
```

```python
ibov
```

```python
ibov.rename(columns={'Adj Close' : 'IBOV'}, inplace = True)
```

```python
ibov = ibov.drop(ibov.columns[[0,1,2,3,5]], axis = 1)
```

```python
ibov
```

```python
PL.index = pd.to_datetime(PL.index)
```

```python
novo_df = pd.merge(ibov,PL, how = 'inner', on = 'Date')
```

```python
novo_df
```

```python
novo_normalizado = novo_df/novo_df.iloc[0]
```

```python
novo_normalizado
```

```python
novo_normalizado[['IBOV','Minha Carteira']].plot(figsize = (10,10));
```
